```python
def job(cols):
    Age=cols[0]
    Pclass= cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
    
        elif Pclass == 2:
            return 29
        
        else:
            return 23
    
    else:
        return Age
    
```

```python
def job3(parch):
    
    if parch==0:
        return 0
    else:
        return 1
```

```python
train['SibSp']=train ['SibSp'].apply(job2)
```

```python
train.head(200)
```

```python
train['Age']= train[['Age','Pclass']].apply(job,axis=1)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False)
```

```python
train.drop('SipSp',axis=1,inplace=True
          )
```

```python
train.dropna(inplace=True)
```

```python
embark = pd.get_dummies(train['Embarked'],drop_first=True)
sex=pd.get_dummies(train['Sex'],drop_first=True)
pclass = pd.get_dummies(train['Pclass'],drop_first=True)
```

```python
train.drop(['Sex','Embarked','Name','Ticket','Pclass'],axis=1,inplace=True)
```

```python
train = pd.concat([train,sex,embark,pclass],axis=1)
```

```python
train.drop('PassengerId',axis=1,inplace=True)
```

```python
train.head()
```

```python
X = train.drop('Survived',axis=1)
y = train['Survived']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=404)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
